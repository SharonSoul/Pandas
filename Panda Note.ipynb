# Pandas
Introduces "Data Frames" and "series" that allow you to slice and dice rows and columns of information.
#### Any depiction of a data set in a table (row and columns) is known as a dataframe
#### Series are gotten from dataframes...each individual column is called a series, e.g age series, name series and so on. A dataframe is  combination of series.
Matplotlib and seabom is used for visual data like graphs, charts and so on.
Scikit_learn is basically for machine learning.

pandas is usually imported into aprogram as .pd  .e.g "import pandas as pd"
numpy is usually imported as np
Matplotlib is usually imported as plt
seabom is usually imported as sns

df is usually used as a variable to equate the pd.method

.head() enables you to view the first five rows, to check what the dataframe being worked with is all about

.tail() enables you to vview the last five rows

.shape specifies the rows and columns i the dataframe
